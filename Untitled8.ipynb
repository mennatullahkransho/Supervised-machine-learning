```python
from sklearn.datasets import load_digits

digits = load_digits()

X = digits.data
y = digits.target
```

```python
import numpy as np

# Filter out all other classes except class 0 and 1
indices = np.where((y == 0) | (y == 1))[0]
X = X[indices]
y = y[indices]

# Convert class labels to binary values
y = np.where(y == 0, -1, y)
y = np.where(y == 1, 1, y)
y = y.reshape(-1, 1)
```

```python
def standardize(X):
    """
    Standardize the dataset X.

    Arguments:
    X -- numpy array of shape (m, n)

    Returns:
    X_std -- standardized numpy array of shape (m, n)
    """
    # Compute the mean and standard deviation of each feature
    mu = np.mean(X, axis=0)
    sigma = np.std(X, axis=0)

    # Subtract the mean and divide by the standard deviation for each feature
    X_std = (X - mu) / sigma

    return X_std

X_std = standardize(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X_std, y, test_size=0.2, random_state=42)
```

```python
class LogisticRegression:
    def __init__(self, num_features):
        self.num_features = num_features
        self.weights = np.zeros((num_features, 1))
        self.bias = 0

    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))

    def predict(self, X):
        z = np.dot(X, self.weights) + self.bias
        return self.sigmoid(z)

    def compute_loss(self, y_true, y_pred):
        eps = 1e-10
        return -np.mean(y_true * np.log(y_pred + eps) + (1 - y_true) * np.log(1 - y_pred + eps))

    def compute_gradients(self, X, y_true, y_pred):
        m = y_true.shape[0]
        dz = y_pred - y_true
        dw = 1/m * np.dot(X.T, dz)
        db = np.sum(dz) / m
        return dw, db

    def update_parameters(self, dw, db, learning_rate, l1_lambda):
        # L1 regularization
        if l1_lambda > 0:
            self.weights -= learning_rate * (dw + l1_lambda * np.sign(self.weights))
        else:
            self.weights -= learning_rate * dw

        self.bias -= learning_rate * db

    def train(self, X, y_true, learning_rate=0.01, num_epochs=100, batch_size=None, l1_lambda=0):
        losses = []
        for epoch in range(num_epochs):
            if batch_size is None:
                # Use entire training set
                X_batch = X
                y_batch = y_true
            else:
                # Sample a minibatch
                idx = np.random.choice(X.shape[0], batch_size, replace=False)
                X_batch = X[idx]
                y_batch = y_true[idx]

            y_pred = self.predict(X_batch)
            loss = self.compute_loss(y_batch, y_pred)
            dw, db = self.compute_gradients(X_batch, y_batch, y_pred)
            self.update_parameters(dw, db, learning_rate, l1_lambda)

            losses.append(loss)

            if epoch % 100 == 0:
                print(f"Epoch {epoch}, Loss: {loss}")
        return losses

    def gradient_descent(self,X_train, y_train, X_val, y_val, lambd, learning_rate, num_iterations):
        # Initialize weights and bias
        n = X_train.shape[1]
        w = np.zeros((n, 1))
        b = 0

        # Initialize lists to store cost function values
        J_train = []
        J_val = []

        # Perform gradient descent for num_iterations
        for i in range(num_iterations):
            # Compute gradients
            dw, db = self.compute_gradient(X_train, y_train, w, b, lambd)

            # Update weights and bias
            w -= learning_rate * dw
            b -= learning_rate * db

            # Compute cost function values
            J_train.append(self.compute_cost(X_train, y_train, w, b, lambd))
            J_val.append(self.compute_cost(X_val, y_val, w, b, lambd))

        return J_train, J_val, w, b

```

```python
learning_rate = 0.01
num_iterations = 1000

# Train logistic regression model with L1 regularization and lambda=0.01
J_train1, J_val1, w1, b1 = gradient_descent(X_train, y_train, X_val, y_val, lambd=0.01, learning_rate=learning_rate, num_iterations=num_iterations)

# Train logistic regression model with L1 regularization and lambda=0.1
J_train2, J_val2, w2, b2 = gradient_descent(X_train, y_train, X_val, y_val, lambd=0.1, learning_rate=learning_rate, num_iterations=num_iterations)
```

```python
def mini_batch_gradient_descent(X_train, y_train, X_val, y_val, lambd, learning_rate, batch_size, num_iterations):
    # Initialize weights and bias
    n = X_train.shape[1]
    w = np.zeros((n, 1))
    b = 0

    # Initialize lists to store cost function values
    J_train = []
    J_val = []

    # Compute number of batches
    m = X_train.shape[0]
    num_batches = int(m / batch_size)

    # Perform mini-batch gradient descent for num_iterations
    for i in range(num_iterations):
        # Randomly shuffle training examples
        permutation = np.random.permutation(m)
        X_train_shuffled = X_train[permutation]
        y_train_shuffled = y_train[permutation]

        # Process each batch of training examples
        for j in range(num_batches):
            # Select a batch of training examples
            start_index = j * batch_size
            end_index = min((j+1) * batch_size, m)
            X_batch = X_train_shuffled[start_index:end_index]
            y_batch = y_train_shuffled[start_index:end_index]

            # Compute gradients on this batch
            dw, db = compute_gradient(X_batch, y_batch, w, b, lambd)

            # Update weights and bias based on gradients
            w -= learning_rate * dw
            b -= learning_rate * db

        # Compute cost function values after processing the entire training set
        J_train.append(compute_cost(X_train, y_train, w, b, lambd))
        J_val.append(compute_cost(X_val, y_val, w, b, lambd))

    return J_train, J_val, w, b


learning_rate = 0.01
num_iterations = 1000

# Train logistic regression model with L1 regularization and lambda=0.01 using mini-batch gradient descent
J_train3, J_val3, w3, b3 = mini_batch_gradient_descent(X_train, y_train, X_val, y_val, lambd=0.01, learning_rate=learning_rate, batch_size=32, num_iterations=num_iterations)

# Train logistic regression model with L1 regularization and lambda=0.1 using mini-batch gradient descent
J_train4, J_val4, w4, b4 = mini_batch_gradient_descent(X_train, y_train, X_val, y_val, lambd=0.1, learning_rate=learning_rate, batch_size=32, num_iterations=num_iterations)
```

```python
def rms_prop(X_train, y_train, X_val, y_val, lambd, learning_rate, decay_rate, num_iterations):
    # Initialize weights and bias
    n = X_train.shape[1]
    w = np.zeros((n, 1))
    b = 0

    # Initialize lists to store cost function values
    J_train = []
    J_val = []

    # Initialize cache for RMS Prop
    s_dw = np.zeros((n, 1))
    s_db = 0

    # Perform RMS Prop for num_iterations
    for i in range(num_iterations):
        # Compute gradients
        dw, db = compute_gradient(X_train, y_train, w, b, lambd)

        # Update cache with exponentially weighted averages of squared gradients
        s_dw = decay_rate * s_dw + (1 - decay_rate) * np.square(dw)
        s_db = decay_rate * s_db + (1 - decay_rate) * np.square(db)

        # Update weights and bias based on gradients and cache
        w -= learning_rate * dw / np.sqrt(s_dw + 1e-8)
        b -= learning_rate * db / np.sqrt(s_db + 1e-8)

        # Compute cost function values
        J_train.append(compute_cost(X_train, y_train, w, b, lambd))
        J_val.append(compute_cost(X_val, y_val, w, b, lambd))

    return J_train, J_val, w, b


def adam(X_train, y_train, X_val, y_val, lambd, learning_rate, beta1, beta2, num_iterations):

    # Initialize weights and bias
    n = X_train.shape[1]
    w = np.zeros((n, 1))
    b = 0

    # Initialize lists to store cost function values
    J_train = []
    J_val = []

    # Initialize first and second moments for Adam
    v_dw = np.zeros((n, 1))
    v_db = 0
    s_dw = np.zeros((n, 1))
    s_db = 0

    # Perform Adam for num_iterations
    for i in range(num_iterations):
        # Compute gradients
        dw, db = compute_gradient(X_train, y_train, w, b, lambd)

        # Update biased first moment estimate with exponentially weighted average of gradients
        v_dw = beta1 * v_dw + (1 - beta1) * dw
        v_db = beta1 * v_db + (1 - beta1) * db

        # Update biased second moment estimate with exponentially weighted average of squared gradients
        s_dw = beta2 * s_dw + (1 - beta2) * np.square(dw)
        s_db = beta2 * s_db + (1 - beta2) * np.square(db)

        # Correct bias in first and second moment estimates
        v_hat_dw = v_dw / (1 - beta1 ** (i+1))
        v_hat_db = v_db / (1 - beta1 ** (i+1))
        s_hat_dw = s_dw / (1 - beta2 ** (i+1))
        s_hat_db = s_db / (1 - beta2 ** (i+1))

        # Update weights and bias based on corrected moment estimates
        w -= learning_rate * v_hat_dw / np.sqrt(s_hat_dw + 1e-8)
        b -= learning_rate * v_hat_db / np.sqrt(s_hat_db + 1e-8)

        # Compute cost function values
        J_train.append(compute_cost(X_train, y_train, w, b, lambd))
        J_val.append(compute_cost(X_val, y_val, w, b, lambd))

    return J_train, J_val, w, b
```

```python
learning_rate = 0.01
num_iterations = 1000

# Train logistic regression model with L1 regularization and lambda=0.01 using RMS Prop
J_train5, J_val5, w5, b5 = rms_prop(X_train, y_train, X_val, y_val, lambd=0.01, learning_rate=learning_rate, decay_rate=0.9, num_iterations=num_iterations)

# Train logistic regression model with L1 regularization and lambda=0.01 using Adam
J_train6, J_val6, w6, b6 = adam(X_train, y_train, X_val, y_val, lambd=0.01, learning_rate=learning_rate, beta1=0.9, beta2=0.999, num_iterations=num_iterations)
```
